```python
steps = [
    [
        SumScaler("both"),
        VectorScaler(target="matrix")
    ],
    [
        ELECTRE2(),
        TOPSIS()
    ]
]
skc.pipelines.SKCCombinatorialPipeline(steps=steps)
```

```python
steps = [pipeline_TOPSIS_tzeng2011multiple, pipeline_ELECTRE2_cebrian2009localizacion, pipeline_ELECTRE2_wang2006]
names = []
for step in steps:
    if isinstance(step, list):
        name = "_".join([type(s).__name__.lower() for s in step])
    else:
        name = type(step).__name__.lower()
    names.append(name)
skc.utils.unique_names(names=names, elements=steps)
```

```python
combined_pipelines = skc.pipelines.mkcombinatorial(pipeline_TOPSIS_tzeng2011multiple, pipeline_ELECTRE2_cebrian2009localizacion, pipeline_ELECTRE2_wang2006)
```
